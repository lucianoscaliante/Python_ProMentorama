1. Faça um programa que leia um arquivo texto contendo um lista de endereços IP e gere um outro arquivo, contendo um relátorio dos endereços IP válidos e inválidos

```python
import os 

def ip_lista(ip_string):
    partes = ip_string.split('.')
    if len(partes) != 4:
        return False
    for parte in partes:
        if not parte.isdigit():
            return False
        parte_in = int(parte)
        if parte_in < 0 or parte_in > 255:
            return False
    return True

if os.path.exists("ips.txt"):
    ips = open("ips.txt", "r")
    lista_ips = ips.read().split("\n")
    
    validos =[]
    invalidos = []
    
    for ip in lista_ips:
        if ip_lista(ip) == True:
            validos.append(ip)
        else:
            invalidos.append(ip)
            
    if len(validos) > 0 or len(invalidos) > 0:
        arq_relatorio = open("relatorio.txt", "wt")
        
        if len(validos) > 0:
            arq_relatorio.write("Endereços válidos:")
            for valido in validos:
                arq_relatorio.write(valido+"\n")
                
        if len(invalidos) > 0:
            arq_relatorio.write("\nEndereços inválidos:\n")
            for invalido in invalidos:
                arq_relatorio.write(invalido+"\n")
        
        arq_relatorio.close()
            
    
```

2. Quando falamos sobre erros e exceções, algumas fontes de informações consideram que no processo de depuração, podemos encontrar três tipos de erro: erro de sintaxe, erro de tempo de execução e erro semântico. Defina o conceito de cada um deles 

A) Erro de sintaxe:
R: dentro da área da computação o erro de sintaxe, ele acontece muito na digitação do programador as vezes esquece um letra com as palavras reservada da linguagem o um ";" no codigo e etc.

b) Erro de tempo de execução:
R: Erro de tempo de execução e quando um software ou um hardware que está com algum problema na execução dele é isso impede o software ou o hardware de se executado.

c) Erro semântico:
R: Erro semântico é um erro que sua logica está correta mas o codigo não faz o que era esperado para ser feito, e esse tipo de erro é geralmente mas dificil de ser indentificado e corrigido
