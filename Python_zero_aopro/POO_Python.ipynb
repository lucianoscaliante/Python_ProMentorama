1) Classe Bola: Crie uma classe que modele uma bola:
a. Atributos: Cor, Circunferência, material
b. Método: trocaCor e mostrarCor

```python
class Bola:
    def __init__(self, cor="unknown", circuferencia=0, material="unknown"):
        self.cor = cor
        self.circuferencia = circuferencia
        self.material = material
        
    def TrocaCor(self):
        trocacor = input("Você deseja mudar a cor da bola {} ? [s/n]".format(self.cor))
        
        trocacor = trocacor[0].lower()
        
        if trocacor == "s":
            novacor = input("\n> Digite a nova cor da bola: ")
            self.cor = novacor
        else:
            print("OK a cor irá continuar a mesma!! {}".format(self.cor))
            
    def MostrarCor(self):
        print("\n A cor da bola é {}".format(self.cor))
        
def Principal():
    bola1 = Bola("Laranja", 6, "Papel")
    
    while True:
        bola1.MostrarCor()
        bola1.TrocaCor()
        
        continuar = input("Deseja continuar ? [s/n]")
        continuar = continuar[0].lower()
        if continuar == "n":
            break
            
    bola1.MostrarCor()
    
Principal()
```

4. Classe Pessoa:Crie uma classe que modele uma pessoa:
a. Atributos:nome, idade, peso e altura 
b. Métodos :Envelhecer, engordar , emagrecer, crescer. OBS: por padrão, a cada ano que nossa pessoa envelhece, sendo a idade dela menos que 21 anos, ela deve crescer 0,5 cm.

```python
class Pessoa:
  def __init__(self, nome, idade, peso, altura):
    self.nome = nome
    self.idade = idade
    self.peso = peso
    self.altura = altura

  def Envelhecer(self, ano):
    self.idade += ano
    if (self.idade < 21):
      self.altura(0.5)

  def Engordar(self, peso):
    self.peso += peso

  def Emagrecer(self, peso):
    self.peso -= peso

  def Crescer(self, altura):
    self.altura += altura

def linhas():
  print('-' * 50)


pessoa = Pessoa("Luciano" , 25, 85, 174)
linhas()
pessoa.Engordar(4)
print(vars(pessoa))
linhas()
pessoa.Emagrecer(7)
print(vars(pessoa))
linhas()
pessoa.Crescer(1)
print(vars(pessoa))
linhas()
pessoa.Envelhecer(2)
print(vars(pessoa))
linhas()
```

6. Classe conta corrent: crie uma classe para implementar um conta corrente. A classe deve possuir os seguintes atributos: numero de conta, nome do correntista e saldo. Os métodos são os seguintes: alterarNome, deposito, e saque: No contrutor, saldo e opicional, com valor defult zero e os demais atributos são obrigatorios.

```python
class Conta:
  def __init__(self, numero, nome, saldo=0):
    self.numero = numero
    self.nome = nome
    self.saldo = saldo

  def setNome(self, nome):
    self.nome = nome

  def deposito(self, valor):
    self.saldo += valor

  def saque(self, valor):
    if self.saldo >= valor:
      self.saldo -= valor
    else:
      print("você não possui saldo suficiente para o saque")

def texto():
  print('-' *50)
  print("Extro da sua conta corrente")


conta = Conta("013", "Luciano", 0)
texto()
print(vars(conta))
texto()
conta.setNome("Carlos")
conta.deposito(5000)
print(vars(conta))
texto()
conta.saque(6000)
print(vars(conta))
texto()
conta.saque(1000)
print(vars(conta))
```
